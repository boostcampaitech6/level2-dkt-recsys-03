### Prediction 분포 확인

코드 실행 전, outputs 폴더에 모델 별 prediction 파일 추가

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_name = ['lgbm', 'xgboost', 'catboost', 'lgcn', 'ffm', 'lqtr', 'gru', 'bert']
data_list = []
for name in data_name:
    data = pd.read_csv(f'./outputs/{name}.csv')
    data_list.append(data)
```

```python
for data, name in zip(data_list, data_name):
    fig, ax = plt.subplots()
    plt.scatter(data['id'], data['prediction'])
    ax.set_title(f'Distribution of Prediction : {name}')
    ax.set_xlabel('Id')
    ax.set_ylabel('Prediction')
    ax.set_ylim(0,1)
    plt.gca().axes.xaxis.set_ticks([])
    ax.axhline(0.5, color='red')
    plt.show()

    print(f"Number of Outlier : {len(data[(data['prediction'] > 1) | (data['prediction'] < 0)])}")
```
